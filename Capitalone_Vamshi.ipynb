Here we are programmatically identifying the reversed and multi-swipe transactions(here created column Duplicated). Here I only counted the transactions which are multiswiped within 720 secs considering the optimal because I tested it with various values from 100 secs to 1000 secs and found this is the best case.

```python
cpone['Duplicated'] = (cpone.sort_values(['ts'])
                       .groupby(['customerId', 'transactionAmount'], sort=False)['ts']
                       .diff()
                       .dt.total_seconds()
                       .lt(720))
```

# Logistic Regression 

```python
cpone_1 = cpone.drop(columns = ['merchantCity','acqCountry', 'merchantName','echoBuffer','merchantState', 'transactionDateTime','merchantZip','cardPresent','posOnPremises','recurringAuthInd','posEntryMode', 'posConditionCode', 'merchantCategoryCode','currentExpDate','accountOpenDate','dateOfLastAddressChange','merchantCountryCode'])
```

```python
cpone_1 = cpone_1.dropna(axis=0)
```

```python
cpone_1['ts'] = cpone_1.ts.astype("category").cat.codes
```

```python
cpone_1.dtypes
```

```python
x_values = cpone_1.drop(['isFraud'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_values, 
                                                    cpone_1[['isFraud']], 
                                                    test_size=0.2, 
                                                    random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
mnlr = LogisticRegression(random_state=0, multi_class='multinomial', solver='newton-cg', penalty = 'l2',
                         C = 2, max_iter = 100)
```

```python
mnlr_model = mnlr.fit(x_train, y_train)
```

```python
y_test_mnlr = y_test.copy()
y_test_mnlr.head()
```

```python
y_test_mnlr['MNLR_CLASS_PREDS'] = mnlr_model.predict(x_test)
y_test_mnlr.shape
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

#confusion_matrix(y_pred_df['isFraud'], y_pred_df['CLASS_PREDS'])
```

```python
accuracy_score(y_test_mnlr['isFraud'],  y_test_mnlr['MNLR_CLASS_PREDS'])
```

# Applying Lightgbm

```python
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size = 0.2, random_state = 1)
```

```python
import lightgbm as lgb
```

```python
lgb_mort_train = lgb.Dataset(x_train, y_train)
lgb_val_train = lgb.Dataset(x_val, y_val)
```

```python
lgb_params = {
    'boosting_type': 'gbdt',
    'objective': 'multiclass',
    'metric': 'multi_logloss',
    'max_depth' : 3,
    'num_class':5,
    'learning_rate': 0.1,
    'early_stopping_round' : 10,
   
}
```

```python
lgb_gbm = lgb.train(params = lgb_params, train_set = lgb_mort_train,
                num_boost_round = 100, valid_sets = [lgb_val_train, lgb_mort_train],
               valid_names = ['Evaluation', 'Train'])
```

```python
y_pred_df_test = y_test.copy()
y_pred_df_train = y_train.copy()
```

```python
y_probs_train = lgb_gbm.predict(x_train)
y_probs_test = lgb_gbm.predict(x_test)
```

```python
y_probs_train.shape
```

```python
lgb_probs_train = pd.DataFrame(lgb_gbm.predict(x_train))
lgb_probs_train.columns = ['prob' + str(col) for col in lgb_probs_train.columns]
lgb_probs_train.head()

lgb_probs_test = pd.DataFrame(lgb_gbm.predict(x_test))
lgb_probs_test.columns = ['prob' + str(col) for col in lgb_probs_test.columns]
lgb_probs_test.head()
```

```python
y_pred_df_test['LG1'] = np.where(y_pred_df_test['isFraud']== 0, 1, 0)
y_pred_df_test['LG2'] = np.where(y_pred_df_test['isFraud']== 1, 1, 0)


y_pred_df_train['LG1'] = np.where(y_pred_df_train['isFraud']== 0, 1, 0)
y_pred_df_train['LG2'] = np.where(y_pred_df_train['isFraud']== 1, 1, 0)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score

plt.plot(roc_curve(y_pred_df_train['LG1'], lgb_probs_train['prob0'])[0],roc_curve(y_pred_df_train['LG1'], lgb_probs_train['prob0'])[1], 
         color = 'blue', label='Rating: 0 ROC Curve (area = %0.5f)' % roc_auc_score(y_pred_df_train['LG1'], lgb_probs_train['prob0']))

plt.plot(roc_curve(y_pred_df_train['LG2'], lgb_probs_train['prob1'])[0],roc_curve(y_pred_df_train['LG2'], lgb_probs_train['prob1'])[1], 
         color = 'red', label='Rating: 1 ROC Curve (area = %0.5f)' % roc_auc_score(y_pred_df_train['LG2'], lgb_probs_train['prob1']))

plt.plot([0, 2], [0, 2], color='black', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Light GBM AUC')
plt.legend()
plt.show()
sns.set(style='white', rc={'figure.figsize':(10,10)})
```

```python
k_model.fit(x_train, y_train, batch_size = 500, epochs = 3, validation_split = 0.2)
```

```python
k_model = Sequential()
k_model.add(Dense(64, input_dim=11, activation='relu'))
k_model.add(Dense(24, input_dim=11, activation='relu'))
k_model.add(Dense(13, input_dim=11, activation='relu'))
k_model.add(Dense(1))
k_model.compile(loss='mean_squared_error', optimizer='adam')
k_model.fit(x_train, y_train, batch_size = 500, epochs = 3, validation_split = 0.2)
```

```python
score = k_model.evaluate(x_test, y_test, batch_size=128)
```

```python
## Example while adding dropout:
from keras.layers import Dense, Dropout, Activation

k_model = Sequential()
k_model.add(Dense(64, input_dim=11, activation='relu'))
k_model.add(Dropout(0.25))
k_model.add(Dense(24, input_dim=11, activation='relu'))
k_model.add(Dense(13, input_dim=11, activation='relu'))
k_model.add(Dense(1))
k_model.compile(loss='mean_squared_error', optimizer='rmsprop')
k_model.fit(x_train, y_train, batch_size = 500, epochs = 3, validation_split = 0.2)
```

```python
## Example of implementing early stopping:

from keras.callbacks import EarlyStopping
callbacks = [EarlyStopping(monitor='val_loss', patience=5)]

k_model = Sequential()
k_model.add(Dense(64, input_dim=11, activation='relu'))
k_model.add(Dense(24, input_dim=11, activation='relu'))
k_model.add(Dense(13, input_dim=11, activation='relu'))
k_model.add(Dense(1))
k_model.compile(loss='mean_squared_error', optimizer='rmsprop')
k_model.fit(x_train, y_train, batch_size = 500, epochs = 100, validation_split = 0.2, callbacks = callbacks)
```

```python
score = k_model.evaluate(x_test, y_test, batch_size=128)
score
```

```python
# Adding dropout throughout the model:

k_model = Sequential()
k_model.add(Dense(64, input_dim=11, activation='relu'))
k_model.add(Dropout(0.50))
k_model.add(Dense(24, input_dim=11, activation='relu'))
k_model.add(Dropout(0.1))
k_model.add(Dense(13, input_dim=11, activation='relu'))
k_model.add(Dense(1))
k_model.compile(loss='mean_squared_error', optimizer='rmsprop')
k_model.fit(x_train, y_train, batch_size = 500, epochs = 100, validation_split = 0.2, callbacks = callbacks)
```
